```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

## Common Discrete Random Variables

### Bernoulli Random Variable
We already made one of these by hand, but there is an easier way:

```python
?stats.bernoulli
```

```python
# probability of 0.2 that the variable is equal to 1
B1=stats.bernoulli(0.2)
```

```python
b=range(-1,3)
```

```python
plt.stem(b,B1.pmf(b))
```

```python
x=np.linspace(-1,2,101)
```

```python
plt.plot(x,B1.cdf(x))
```

```python
num_sims=100000
b=B1.rvs(size=num_sims)
np.sum(b==1)/num_sims
```

```python
mybins=[0,1,2]
plt.hist(b,bins=mybins,density=True);
```

```python
plt.hist(b,density=True);
```

```python
mybins=[0,1,2]
plt.hist(b,bins=mybins,density=True, alpha=0.25)
plt.stem(mybins,B1.pmf(mybins))
```

## Binomial Random Variable

```python
?stats.binom
```

```python
# How many successes on 12 Bernoulli Trials when the probability of one success is 0.2?
# This is the exact same as the sum of 12 Bernoulli trials together
Bn=stats.binom(12,0.2)
```

```python
x=range(0,13)
plt.stem(x,Bn.pmf(x))
```

```python
x=np.linspace(-5,15,1000)
plt.plot(x,Bn.cdf(x))
```

```python
p=0.2

X=Bn.rvs(size=100000)
mybins=range(0,14)
plt.hist(X,bins=mybins,density=True,alpha=0.3)
```

```python
p=0.7
Bn=stats.binom(12,p)
bn=Bn.rvs(size=10000)
mybins=range(0,14)
plt.hist(bn,bins=mybins,density=True,alpha=0.3)

x=range(0,13)
plt.stem(x,Bn.pmf(x))
```

```python
p=0.2
Bn=stats.binom(12,p)


X=Bn.rvs(size=10000)
mybins=range(0,14)
plt.hist(X,bins=mybins,density=True,alpha=0.3,cumulative=True)

```

```python
p=0.2
Bn=stats.binom(12,p)


bn=Bn.rvs(size=10000)
mybins=range(0,14)
plt.hist(bn,bins=mybins,density=True,alpha=0.3,cumulative=True)

x=np.linspace(-1,13,100)
plt.step(x,stats.binom.cdf(x,12,p)) # Show that replacing with plt.step is better
```

### Sums of BERNOULLI RVs

```python
b12=B1.rvs(size=(100000,12))
bsum=np.sum(b12,axis=1)
mybins=range(0,14)
plt.hist(bsum,bins=mybins,density=True,alpha=0.3)
```

```python
b12=B1.rvs(size=(100000,12))
bsum=np.sum(b12,axis=1)
mybins=range(0,14)
plt.hist(bsum,bins=mybins,density=True,alpha=0.3)

x=range(0,13)
plt.stem(x,Bn.pmf(x))
```

Conclusion: Adding together independent Bernoulli RVs (with the same p) produces a Binomial RV

## Geometric Random Variable

```python
# How many trials does it take until we achieve a success?
# The success of one individual trial has a probability of 0.2.
G=stats.geom(0.2)
```

```python
g=G.rvs(100000)
```

```python
plt.scatter(range(len(g)),g)
```

```python
vals=range(20)
plt.hist(g,bins=vals,density=True,alpha=0.5)
plt.stem(vals,G.pmf(vals))
```

```python
g=G.rvs(1000000)
vals=range(20)
plt.hist(g,bins=vals,density=True,alpha=0.5,cumulative=True)
plt.step(vals,G.cdf(vals),where='post')
```

If you flip a coin until you see heads, what is the probability that it takes more than 6 flips?

$P(G>6)=1- P(G \le 6)=1-F_X(6)$

```python
G=stats.geom(0.5)
```

```python
1-G.cdf(6)
```

Write a simulation to verify this:

```python
num_sims=1000000
count=0
for sim in range(num_sims):
    coin=npr.randint(0,2,6)
    if coin.sum()==0:
        count+=1
        
print(count/num_sims)
        
```
