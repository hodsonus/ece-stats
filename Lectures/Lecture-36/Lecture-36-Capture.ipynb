```python
import numpy as np
import numpy.random as npr
import scipy.stats as stats
```

```python
handed=np.array([[43,9],
                [44,4]])
```

```python
stats.contingency.expected_freq(handed)
```

```python
stats.fisher_exact(handed)
```

```python
Eh=stats.contingency.expected_freq(handed)
```

```python
Cvals=(handed-Eh)**2/Eh
Cvals
```

```python
C=np.sum(Cvals)
C
```

```python
stats.chi2.sf(C,1)
```

```python
?stats.chi2_contingency
```

```python
stats.chi2_contingency(handed,correction=False)
```

```python
stats.chi2_contingency(handed,correction=True)
```

```python
wot=pd.crosstab(gss["DEGREE"],gss["WOTRIAL"])
```

```python
wot
```

```python
wot=wot.drop([0,8,9],axis=1)
wot
```

```python
wot=wot.drop([9])
wot
```

```python
wot.columns=['Definitely', 'Probably', "Probably Not", 'Definitely Not']
wot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for i in wot.columns:
    plt.plot(wot.index, wot[i], label=i)
plt.legend()
```

```python
wot.index=['Less than HS', 'HS', 'JuCo', 'Bachelors', 'Graduate']
wot
```

```python
print("Expected Contingency Table")
Ewot=stats.contingency.expected_freq(wot)
Ewot
```

```python
print("Differences from Expected")
wot-Ewot
```

```python
stats.contingency.chi2_contingency(wot)
```

The result is statisticaly significant: the row and columns are not independent (p<0.01)
