```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
?stats.norm 
#Gaussian RV
# survival fxn, 1-cdf
# if we were makign a normal fxn, this would be the Q fxn
```

```python
# mean - "where am I centered"
# variance - "how far are my points distributed"
# area under curve - 1


G1=stats.norm(0,1) # mean of 0, centered at 0. variance of 1

G2=stats.norm(10,3) # mean of 10, centered at 10

G3=stats.norm(-5,0.3) # mean of -5, centered at -5
```

```python

x=np.linspace(-8,18,1000)
plt.plot(x,G1.pdf(x),label='norm(0,1)')
plt.plot(x,G2.pdf(x),'--',label='norm(10,3)')
plt.plot(x,G3.pdf(x),'--',label='norm(-5,0.3)')
plt.legend()
```

```python
X = G1.rvs(size=10000)
```

```python
plt.hist(X,bins=30,density=True);
plt.plot(x,G1.pdf(x));
plt.xlim(-5,5)
```

```python
x=np.linspace(-8,18,1000)
plt.plot(x,G1.cdf(x),label='norm(0,1)')
plt.plot(x,G2.cdf(x),'--',label='norm(10,3)')
plt.plot(x,G3.cdf(x),'--',label='norm(-5,0.3)')
plt.legend()
```
