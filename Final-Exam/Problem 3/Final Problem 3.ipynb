```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import scipy.stats as stats
import sympy as sp
```

```python
muA = 1
muB = 2
A=stats.expon(scale=muA)
B=stats.expon(scale=muB)
```

```python
def plot(probA):
    x=np.linspace(-1,15,100)
    probB = 1-probA
    plt.plot(x,probA*A.pdf(x), label="A")
    plt.plot(x,probB*B.pdf(x), label="B")
    plt.legend()
```

```python
print("(a)")
probA = 0.7
plot(probA)

# this is very close to what I want to do with the solving, couldnt get it working :(
# xa=sp.symbols('x')
# print(sp.solve(probA*sp.functions.elementary.exponential(-1*xa)-probB*1/2*sp.functions.elementary.exponential(-1/2*xa)))
```
